#### basically quantum-ify a pic of a lemur

```python
import numpy as np

import quantumblur as qb
from PIL import Image
```

```python
ac = Image.open('lemur.jpg')

qb.circuits2image(qb.image2circuits(ac))
```

```python
def partial_x(qc, fraction):
    for i in range(qc.num_qubits):
        qc.rx(np.pi * fraction, i)
```

```python
lemur_circuits = qb.image2circuits(ac)

for circuit in lemur_circuits:
    partial_x(circuit, 0.2)
    
rotated_lemur = qb.circuits2image(lemur_circuits)

rotated_lemur
```
